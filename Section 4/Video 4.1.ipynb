```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Generate some mock data
from scipy.ndimage.filters import gaussian_filter
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
z = np.random.normal(size=1000)
im = gaussian_filter(np.random.normal(size=(512,512)), 16)
```

```python
#Show available stylesheets
plt.style.available
```

```python
# Stylesheets
plt.style.use('ggplot')
plt.subplot(221)
plt.plot(np.arange(0, 20, 0.1),np.sin(np.arange(0, 20, 0.1)), 'o-')
plt.subplot(222)
plt.scatter(x,y, c=z)
plt.subplot(223)
plt.hist((x,y,z))
plt.subplot(224)
plt.imshow(im)
plt.gcf().set_size_inches(12,8)
```

```python
# LaTeX Labels
plt.hist((-x**2,-y**2,-z**2));
plt.xlabel('x')
plt.ylabel(r'$\nabla\dot x^2$')
```

```python
# Font properties
plt.hist((-x**2,-y**2,-z**2));
plt.xlabel('x')
plt.ylabel('x^2')
```

```python
# Tweaking the legend
plt.hist((-x**2,-y**2,-z**2), label=('x','y','z'));
plt.legend(framealpha=0.5)
plt.xlabel('x')
plt.ylabel('x^2')
```
