```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
def setup_map(m):
    m.drawcoastlines(zorder=-1)
    m.fillcontinents(color='olive',lake_color='aqua', zorder=-1)
    # draw parallels and meridians.
    m.drawmapboundary(fill_color='aqua')
    plt.gcf().set_size_inches(12,10)
```

```python
# What is the shortest distance between two points?
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
m.plot([-10, 60], [45, 45], 'ko-', latlon=True, lw=3)
m.plot(np.linspace(-10,60,20), 45*np.ones(20), 'ro-', latlon=True, lw=3)
setup_map(m)
```

```python
# The great circle!
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
m.plot([-10, 60], [45, 45], 'ko-', latlon=True, lw=3)
m.plot(np.linspace(-10,60,20), 45*np.ones(20), 'ro-', latlon=True, lw=3)
m.drawgreatcircle(-10, 45, 60, 45, linestyle='--', lw=3)
setup_map(m)
```

```python
# All longitude lines are great circles
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
m.drawgreatcircle(-10, 45, 60, 45, linestyle='--', lw=3)
m.plot(20*np.ones(20), np.linspace(20,60,20), 'ro-', latlon=True, lw=3)
m.drawgreatcircle(20, 20, 20, 60, lw=3, linestyle='--')
m.plot([20,20], [20, 60], 'ko-', lw=3, latlon=True)
setup_map(m)
```

```python
# Not all projections show great circles as straight lines!
#m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
m = Basemap(projection='cyl', llcrnrlat=15,urcrnrlat=75, llcrnrlon=-30,urcrnrlon=70, resolution='l')
m.plot(np.linspace(-10,60,20), 45*np.ones(20), 'ro-', latlon=True, lw=3)
m.drawgreatcircle(-10, 45, 60, 45, linestyle='--', lw=3)
setup_map(m)
```
